**Task 1 Simple Task List**

Description : Develop a Python task management app for adding, removing, listing, prioritizing, and receiving task recommendations based on task descriptions.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
import random
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Initialize an empty task list
tasks = pd.DataFrame(columns=['description', 'priority'])
```

```python
# Load pre-existing tasks from a CSV file (if any)
try:
    tasks = pd.read_csv('tasks.csv')
except FileNotFoundError:
    pass
```

```python
# Function to save tasks to a CSV file
def save_tasks():
    tasks.to_csv('tasks.csv', index=False)
```

```python
# Train the task priority classifier
vectorizer = CountVectorizer()
clf = MultinomialNB()
model = make_pipeline(vectorizer, clf)
model.fit(tasks['description'], tasks['priority'])
```

```python
tasks = []
# Function to add a task to the list
def add_task(description, priority):
    tasks.append({"description": description, "priority": priority})
```

```python
# Function to remove a task by description
def remove_task(description):
    global tasks
    tasks = [task for task in tasks if task["description"] != description]
```

```python
# Function to list all tasks
def list_tasks():
    clear_output(wait=True)
    if tasks:
        print("Task List:")
        for task in tasks:
            print(f"- {task['description']} ({task['priority']})")
    else:
        print("No tasks available.")
```

```python
# Function to recommend a task based on machine learning
def recommend_task():
    clear_output(wait=True)
    if tasks:
        task = sorted(tasks, key=lambda t: t["priority"])[0]
        print(f"Recommended task: {task['description']} ({task['priority']})")
    else:
        print("No tasks to recommend.")
```

```python
# Widget Setup
desc_input = widgets.Text(placeholder='Task description', description='Task:')
prio_dropdown = widgets.Dropdown(options=['Low', 'Medium', 'High'], description='Priority:')
add_button = widgets.Button(description="Add Task")
remove_input = widgets.Text(placeholder='Task description to remove', description='Remove:')
remove_button = widgets.Button(description="Remove Task")
list_button = widgets.Button(description="List Tasks")
recommend_button = widgets.Button(description="Recommend Task")
```

```python
# Callbacks
def on_add_clicked(_):
    add_task(desc_input.value, prio_dropdown.value)
    desc_input.value = ""
    list_tasks()

def on_remove_clicked(_):
    remove_task(remove_input.value)
    remove_input.value = ""
    list_tasks()

def on_list_clicked(_):
    list_tasks()

def on_recommend_clicked(_):
    recommend_task()
```

```python
# Assign callbacks
add_button.on_click(on_add_clicked)
remove_button.on_click(on_remove_clicked)
list_button.on_click(on_list_clicked)
recommend_button.on_click(on_recommend_clicked)
```

```python
# Display Widgets
display(desc_input, prio_dropdown, add_button, remove_input, remove_button, list_button, recommend_button)
```
